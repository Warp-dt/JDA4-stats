```python
def sum_dict(d1,d2):
    return {k: d1.get(k, 0) + d2.get(k, 0) for k in set(d1) | set(d2)}
```

```python
def subsets(s):
    sl=list(s)
    return [{sl[0],sl[1]},{sl[0],sl[2]},{sl[1],sl[2]}]

subsets({'Herdegrize', 'Oshimo', 'Terra Cogita'})   
```

```python
def dict_count_val_over(dic,limit):
    count=0
    for v in dic.values():
        if v>limit:
            count+=1
    return count
```

```python
def dict_barh(dic,title="michel",sort=False):
    
    if sort:
        dic_s=dict(sorted(dic.items(), key=lambda item: item[1]))

    else:
        dic_s=dic
    compp=dic_s.items()
    x, y = zip(*compp)   
    # print(x,y)
    offset_text=(max(y)/60)

    fig, ax=plt.subplots(figsize=(10,5),facecolor='#7B7826')
    ax.set_facecolor("#B69E44")

    x_strip=[k.replace("{","").replace("}","").replace('"',"").replace("'","") for k in x]
    ax.barh(x_strip, y,facecolor='#231D0F')
    for i in range(len(x)):
        ax.text( y[i]+offset_text,i, y[i], ha='left',va="center",color='black')

    if title!="michel":
        ax.set_title(title)
    plt.tight_layout()
    plt.savefig("graphs_tournament_phase/"+title+'.png')
    
    plt.show()
```

```python
#fonction qui sert Ã  filtrer et enlever les keys dont la valeur est en dessous de 3

def filterTheDict(dictObj, callback):
    newDict = dict()
    # Iterate over all the items in dictionary
    for (key, value) in dictObj.items():
        # Check if item satisfies the given condition then add to new dict
        if callback((key, value)):
            newDict[key] = value
    return newDict
```

```python
print(d[0]["players"])

def extract_play(s):
    spl=s.split('title=\\"')

    fi=spl[1].split('\\\"')[0]
    sec=spl[2].split('\\\"')[0]
    ter=spl[3].split('\\\"')[0]

    spl2=s.split('href=\\"/fr/joueur/')
    fi_id=spl2[1].split('\\"')[0]
    sec_id=spl2[2].split('\\"')[0]
    ter_id=spl2[3].split('\\"')[0]
    return fi,sec,ter,fi_id,sec_id,ter_id

extract_play(d[0]["players"])
```

```python
print(d[0]["name"])

def extract_team(s):
    # id_team=int(re.findall("[0-9]+",s)[0])
    
    name=s.split("'>")[1][:-4]

    return name#,id_team


extract_team(d[0]["name"])
```

```python
print(d[0]["classes"])

def extract_classes(s):
    spl=s.split("title='")[1:]
    spl2=tuple([ ss.split("' ")[0]  for ss in spl])
    # sec=spl[1][:-4]
    # fi=spl[0].split('"')[-1]
    return spl2

extract_classes(d[0]["classes"])
```

```python
p1=[]
p2=[]
p3=[]

p1_id=[]
p2_id=[]
p3_id=[]

c1=[]
c2=[]
c3=[]

team=[]

for dic in d:
    (tp1,tp2,tp3,tp1_id,tp2_id,tp3_id)=extract_play(dic["players"])
    p1.append(tp1)
    p2.append(tp2)
    p3.append(tp3)

    p1_id.append(tp1_id)
    p2_id.append(tp2_id)
    p3_id.append(tp3_id)

    (tc1,tc2,tc3)=extract_classes(dic["classes"])
    c1.append(tc1)
    c2.append(tc2)
    c3.append(tc3)
    
    team.append(extract_team(dic["name"]))
    
```

```python
df=pd.DataFrame()
df["equipes"]=team

df["J1"]=p1
df["J2"]=p2
df["J3"]=p3

df["ID_J1"]=p1_id
df["ID_J2"]=p2_id
df["ID_J3"]=p3_id

df["classe_J1"]=c1
df["classe_J2"]=c2
df["classe_J3"]=c3

df=df.set_index("equipes")
df
```

```python
tab_raw=big_element[-1]
tab_str=str(tab_raw.string)
# print(tab_str)
trim=tab_str[29:-830]
n=len(trim)
print(trim[:100])
print(trim[n-100:])

d=json.loads(trim)
matches_raw=[didi["html"] for didi in d]
no_score=matches_raw[0]
w_score=matches_raw[-1]
```

```python
def extract_match_not_played(s):
    spl=s.split('title=\\"')
    team1=spl[1].split(" - ")[0]
    team2=spl[3].split(" - ")[0]

    if team1=="- Micans":
        team1="- Micans -"
    if team2=="- Micans":
        team2="- Micans -"
    if team1=="- Phoenix":
        team1="- Phoenix -"
    if team2=="- Phoenix":
        team2="- Phoenix -"
    return team1,team2,"en_attente"
```

```python
def extract_match_played(s):
    spl=s.split('title=\\"')
    # print(spl)
    team1=spl[1].split(" - ")[0]
    team2=spl[3].split(" - ")[0]

    if len(re.findall("score win",spl[1]))>0:
        team_win=team1
    else:
        team_win=team2

    if team1=="- Micans":
        team1="- Micans -"
    if team2=="- Micans":
        team2="- Micans -"
    if team1=="- Phoenix":
        team1="- Phoenix -"
    if team2=="- Phoenix":
        team2="- Phoenix -"
    if len(team1)>20:
        team1=team1.split('\\">')[0]
    if len(team2)>20:
        team2=team2.split('\\">')[0]
    return team1,team2,team_win
```

```python
team_left=[]
team_right=[]
vainqueur=[]

for s in matches_raw:

    if len(re.findall("score win",s))==0: #match en attente d'etre jouÃ©
               
        temp_left,temp_right,temp_win=extract_match_not_played(s)

        if len(re.findall("score forfeit",s))>1:
            temp_win="double_forfait"
        if len(re.findall("score draw",s))>1:
            temp_win='draw'

    else: #match jouÃ©
        temp_left,temp_right,temp_win=extract_match_played(s)
        # if '-' in temp_left or '-' in temp_right or '-' in temp_win:
        #     print(s)
    
    team_left.append(temp_left)
    team_right.append(temp_right)
    vainqueur.append(temp_win)


```
